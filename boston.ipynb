```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats #과학용 계산 라이브러리
from sklearn.datasets import load_boston

boston=load_boston()
```

```python
data = pd.DataFrame(boston.data, columns = boston.feature_names)
data
```

```python
data['PRICE']=boston.target
```

```python
#EDA

fig, axs = plt.subplots(figsize=(16,8), ncols=4, nrows=2)

lm_features = ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']

for i, feature in enumerate(lm_features):
    row = int(i/4)
    col = i%4

    sns.regplot(x=feature, y='PRICE', data=data, ax=axs[row][col])
```

```python
#상관 관계 행렬
df_corr=data.corr()

#히트맵
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()
```

```python
#LSTAT, RM, PTRATIO, INDUS만 추출
df=data.drop(['CRIM', 'ZN', 'CHAS', 'NOX','AGE', 'DIS', 'RAD', 'TAX', 'B','PRICE'],axis=1)
df.head()
```

```python
#StandScaler()
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler=scaler.fit_transform(df)
scaler=pd.DataFrame(df, columns=df.columns)
```

```python
sns.boxplot(data=scaler)
plt.show()
```

```python
#예측

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X=df.loc[:, ['RM','INDUS','PTRATIO']]
y=df.loc[:, 'LSTAT']

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2,
                                                  shuffle=True, random_state=0)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print("회귀계수(기울기):", np.round(lr.coef_, 1)) 
print("상수항(절편):", np.round(lr.intercept_, 1)) 
```

```python
y_preds = lr.predict(X_test)

print(X_train.shape, X_test.shape)
```

```python
print("절편 값:", lr.intercept_)

print("회귀계수:", np.round(lr.coef_,1))
```

```python
mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
print(f'MSE : {mse:.3f}, RMSE: {rmse:.3f}')
print(f'Variance score : {r2_score(y_test, y_preds):.3f}')
```

```python
#실제 데이터와 예측 데이터를 산점도 표시(X의 각 피처별로 산점도 표시)

#RM
y_test_pred=lr.predict(X_test)

plt.figure(figsize=(10,5))
plt.scatter(X_test['RM'], y_test, label='y_test')
plt.scatter(X_test['RM'], y_test_pred, c='r', label='y_pred')

plt.show() 
```

```python
#INDUS
y_test_pred = lr.predict(X_test)

plt.figure(figsize=(10,5))
plt.scatter(X_test['INDUS'], y_test, label='y_test')
plt.scatter(X_test['INDUS'], y_test_pred, c='r', label='y_pred')

plt.show()  
```

```python
#PTRATIO
y_test_pred = lr.predict(X_test)

plt.figure(figsize=(10,5))
plt.scatter(X_test['PTRATIO'], y_test, label='y_test')
plt.scatter(X_test['PTRATIO'], y_test_pred, c='r', label='y_pred')

plt.show()  
```
